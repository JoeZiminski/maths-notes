## SOH CAH TOA

Relate the angle on a right-angled triangle to the length of the triangle's sides:

$$
\begin{gather*}
    \sin(\theta) = \frac{\textnormal{|opposite|}}{\textnormal{|hypotenuse|}} \\[12pt]
    \cos(\theta) = \frac{\textnormal{|adjacent|}}{\textnormal{|hypotenuse|}} \\[12pt]
    \tan{\theta} = \frac{\textnormal{|opposite|}}{\textnormal{|adjacent|}}
\end{gather*}
$$

## Sine Rule

Capital letters are an angle, small letters are a side.

$$
\begin{gather*}
    \frac{a}{\sin(A)} = \frac{b}{\sin(B)} = \frac{c}{\sin(C)} \\[12pt]
    \textnormal{and for finding an unkown angle} \\[12pt]
    \frac{\sin(A)}{a} = \frac{\sin(B)}{b} = \frac{\sin(C)}{c}
\end{gather*}
$$

**When using the sine rule to find an angle, it is essential to check whether an obtuse angle is possible.**

## Cosine Rule

$$
\begin{gather*}
    a^2 = b^2 + c^2 - 2bc\cos(A) \\[4pt]
    \textnormal{similarly...} \\[4pt]
    b^2 = a^2 + c^2 - 2ac\cos(B) \\[4pt]
    c^2 = a^2 + b^2 - 2ab\cos(C)
\end{gather*}
$$

and we can rearrange to express the angle as a function of the sides.

# Trigonometric Functions

**Cosine**

Cosine is an even function. $\cos(\theta) = \cos(-\theta)$

**Sine**

Sine is an odd function. $-\sin(\theta) = \sin(-\theta)$

**Tan**

Tan is odd, $-\tan(\theta) = \tan(-\theta)$.

**Note**

$$
\begin{gather*}
    -\cos(\theta) = \cos(\pi - \theta) \\[4pt]
    \sin(\theta) = \sin(\pi - \theta) \\[4pt]
    -\tan(\theta) = \tan(\pi - \theta)
\end{gather*}
$$

i.e. $\sin$ symmetric about $\dfrac{\pi}{2}$, $\cos$ and $\tan$ are antisymmetric about $\dfrac{\pi}{2}$. 

$\sin$$ and $\cos$ have a period $2\pi$, $\tan$ has a period $\pi$. $\tan$ is undefined at $..., -\dfrac{\pi}{2},\ \dfrac{\pi}{2},\ \dfrac{3\pi}{2}, ...$

## Radians

On a circle, an arc with its length length the same as radius subtends an angle of 1 *radian*. One radian is a little less than $60^\circ$.

$C = 2\pi r$, so on the unit circle the circumference is $2 \pi$. Therefore, $2\pi  \ \textnormal{rad} = 360^\circ$ and $\pi \ \textnormal{rad} = 180^\circ$

In general:

$$
1^\circ = \frac{\pi}{180} \ \textnormal{rad}
$$

e.g. to convert $253^\circ$ to radians: $253^\circ = \frac{253\pi}{180}$. 

To convert from radians to degrees, multiply by $\dfrac{180^\circ}{\pi}$.

## Inverse Trig Functions

$$
\begin{gather*}
    \theta = \arcsin\bigg(\frac{\textnormal{opposite}}{\textnormal{hypotenuse}}\bigg) \\[12pt]
    \theta = \arccos\bigg(\frac{\textnormal{adjacent}}{\textnormal{hypotenuse}}\bigg) \\[12pt]
    \theta = \arctan\bigg(\frac{\textnormal{opposite}}{\textnormal{adjacent}}\bigg)
\end{gather*}
$$

where opposite, adjacent, and hypotenuse denote the lengths of the corresponding sides.

### Reciprocal Trig Identities

$$
\tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)}
$$
$$
\mathrm{cosec}(\theta) = \frac{1}{\sin(\theta)}
$$
$$
\mathrm{sec}(\theta) = \frac{1}{\cos(\theta)}
$$
$$
\mathrm{cot}(\theta) = \frac{\cos(\theta)}{\sin(\theta)}
$$

### Pythagorean Identities

$$
\cos^2(\theta) + \sin^2(\theta) = 1
$$
$$
1 + \tan^2(\theta) = \mathrm{sec}^2(\theta)
$$
$$
\mathrm{cot}^2(\theta) + 1 = \mathrm{cosec}^2(\theta)
$$

From the first identity, we can express sin in terms of cosine as
$\sin^2(\theta) = 1 - \cos^2(\theta)$ and vice versa.

### Compound Angles

It's nice to think about the relationship between taking a function
of the sum of two *angles*, or a sinusoidal function of the sum of two *frequencies.

$$
\sin(A+B) = \sin(A)\cos(B) + \cos(A)\sin(B)
$$
$$
\sin(A-B) = \sin(A)\cos(B) - \cos(A)\sin(B)
$$
$$
\cos(A+B) = \cos(A)\cos(B) - \sin(A)\sin(B)
$$
$$
\cos(A-B) = \cos(A)\cos(B) + \sin(A)\sin(B)
$$
$$
\tan(A + B) = \frac{\tan(A) + \tan(B)}{1 - \tan(A)\tan(B)}
$$
$$
\tan(A - B) = \frac{\tan(A) - \tan(B)}{1 + \tan(A)\tan(B)}
$$

### Difference of Compound Angles

This can be interpreted as adding sines and cosine functions shifted in frequency some factor $B$, expressed as multiple of sine and cosines.

$$
\sin(A+B) + \sin(A-B) = 2\sin(A)\cos(B)
$$
$$
\sin(A+B) - \sin(A-B) = 2\cos(A)\sin(B)
$$
$$
\cos(A+B) + \cos(A-B) = 2\cos(A)\cos(B)
$$
$$
\cos(A+B) - \cos(A-B) = -2\sin(A)\sin(B)
$$


### Double Angle

These are all very useful.

**Cos**

$$
\begin{gather*}
     \cos(2\theta) = 1 - 2\sin^2(\theta) \\[4pt]
     \cos(2\theta) = 2\cos^2(\theta) - 1  \\[4pt]
     \textnormal{rearranging...} \\[4pt]
     \cos^2(\theta) = \frac{1}{2}(1 + \cos(2\theta)) \\[4pt]
     \sin^2(\theta) = \frac{1}{2}(1 - \cos(2\theta))
\end{gather*}
$$

This is a cool one that I always forget:

$$
\begin{gather*}
     \cos(2\theta) = \cos^2(\theta) - \sin^2(\theta)
\end{gather*}
$$

**Sin**
$$
\sin(2\theta) = 2\sin(\theta)\cos(\theta)
$$

**Tan**
$$
\tan(2\theta) = \frac{2\tan(\theta)}{1 - \tan^2(\theta)}
$$

It is nice to remember with these identities that we can switch between cosine and sine interpretation by halving our $\theta$, i.e.

$$
\cos(\theta) = 1 - 2\sin^2\left(\frac{\theta}{2}\right)
$$


## Expressing cos + sin as a shifted sine wave

The sum of a sine and cosine of the same frequency is a shifted sine wave! This is exactly how phase is determined in the Fourier transform!

$$
\begin{gather*}
    M\sin(\theta + \alpha) = a\sin(\theta) + b\cos(\theta) \\[4pt]
    M = \sqrt{a^2 + b^2} \\[4pt]
    \alpha = \tan^{-1}\left(\frac{b}{a} \right) \\[4pt]
\end{gather*}
$$

Note that these are very similar to calculating the angle and magnitude of complex numbers on an Argand diagrams. 
Similar to complex numbers, the quadrant of the angle needs to be checked as $\arctan$ gives only the principal value.

I like the derivation, it uses the angle addition formula:

$$
\begin{gather*}
    M\sin(\theta + \alpha) = M\sin(\theta)\cos(\alpha) + M\cos(\theta)\sin(\alpha) \\[4pt]
    \textnormal{matching terms...} \\[4pt]
    a = M\cos(\alpha) \\[4pt]
    b = M\sin(\alpha) 
\end{gather*}
$$

# Calculus on Trig Functions

## Differentiation

Note the double angle identities are very useful when integrating and differentiating.

$$
\frac{d}{dx}\sin(x) = \cos(x)
$$
$$
\frac{d}{dx}\cos(x) = -\sin(x)
$$

For multiples of sin and cos, the double-angle identities can be very useful. e.g.

$$
\frac{d}{dx}\sin(x)\cos(x) = \dfrac{1}{2} \frac{d}{dx} \sin(2x) = \cos(2x)
$$
