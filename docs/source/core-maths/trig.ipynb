# Triangle Rules

We can use $\sin$, $\cos$ and $\tan$ functions to calculate the sides of triangles. See [here](https://www.quora.com/How-were-trig-functions-calculated-approximated-before-power-series) for an interesting discussion on how these functions are derived.


## SOH CAH TOA

\begin{gather*}
    \sin(\theta) = \frac{\textnormal{opposite}}{\textnormal{hypotenuse}} \\[16pt]
    \cos(\theta) = \frac{\textnormal{adjacent}}{\textnormal{hypotenuse}} \\[16pt]
    \tan{\theta} = \frac{\textnormal{opposite}}{\textnormal{adjacent}}
\end{gather*}

## Sine Rule

Capital letters are an angle. Small letters are a side.

\begin{gather*}
    \frac{a}{\sin(A)} = \frac{b}{\sin(B)} = \frac{c}{\sin(C)} \\[16pt]
    \textnormal{and for finding an unkown angle} \\[16pt]
    \frac{\sin(A)}{a} = \frac{\sin(B)}{b} = \frac{\sin(C)}{c}
\end{gather*}

**When using the sine rule to find an angle, it is essential to check whether an obtuse angle is possible.**

## Cosine Rule

\begin{gather*}
    a^2 = b^2 + c^2 - 2bc\cos(A) \\[16pt]
    \textnormal{similarly..} \\[16pt]
    b^2 = a^2 + c^2 - 2ac\cos(B) \\[16pt]
    \textnormal{and} \\[16pt]
    c^2 = a^2 + b^2 - 2ab\cos(C) \\
\end{gather*}

and we can rearrange to express the angle as a function of the sides.

# Trigonometric Functions

**Cosine**

Cosine is an even function. $\cos(\theta) = \cos(-\theta)$

**Sine**

Sine is an odd function. $-\sin(\theta) = \sin(-\theta)$

**Tan**

Tan is odd, $-\tan(\theta) = \tan(-\theta)$.

**Note**

In the book, they make the point:
$$
-\cos(\theta) = \cos(\pi - \theta),\quad \sin(\theta) = \sin(\pi - \theta),\quad -\tan(\theta) = \tan(\pi - \theta)
$$
This doesn't seem particularly interesting, vaguely interesting that cosine is positive for acute angle and negative for obtuse.

## Radians

In the unit circle, an arc of length 1 radius subtends an angle of 1 radian. One radian is a little less than $60 \degree$, this helps to visualise the size of a radian.

$C = 2\pi r$, so on the unit circle the circumference is $2 \pi$. So $2\pi  \ \textnormal{rad} = 360 \degree$ and $\pi \ \textnormal{rad} = 180 \degree$

In general:

$$
1 \degree = \frac{\pi}{180} \ \textnormal{rad}
$$

e.g. convert $253 \degree$ to radians, $253 \degree = \frac{253\pi}{180}$. For converting from radians to degrees, we multiply by $180\degree$.

## Graphs and the Unit Circle

When plot, $A\sin(\theta_f + \phi)$ is a sine wave with amplitude $A$, frequency $\theta_f$, and phase shift $\phi$.

The sides of a triangle with hypotenuse 1 (e.g. on the unit circle) can be expressed in terms of sin, cos and tan:

$$
\sin(\theta) = y
$$
$$
\cos(\theta) = x
$$
$$
\tan(\theta) = \frac{y}{x}
$$

Similar to the SOH CAH TOA, we can use the inverse trig functions to express equations in terms of $\theta$:

$$
\quad \theta  = \sin^{-1}(y) 
$$
$$
\quad \theta = \cos^{-1}(x) 
$$
$$
\quad \theta = \tan^{-1}\left(\frac{y}{x}\right)
$$

Sin and cos have a period $2\pi$, tan has a period $\pi$. Tan is undefined at $-\dfrac{\pi}{2},\ \dfrac{\pi}{2},\ \dfrac{3\pi}{2}$

### Inverse Trig Identities

$$
\tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)}
$$
$$
\mathrm{cosec}(\theta) = \frac{1}{\sin(\theta)}
$$
$$
\mathrm{sec}(\theta) = \frac{1}{\cos(\theta)}
$$
$$
\mathrm{cot}(\theta) = \frac{\cos(\theta)}{\sin(\theta)}
$$

### Pythagorean Identities

It is very useful to remember: from the first identity, we can express sin in terms of cosine as
$\sin^2(\theta) = 1 - \cos^2(\theta)$ and vice versa.

$$
\cos^2(\theta) + \sin^2(\theta) = 1
$$
$$
1 + \tan^2(\theta) = \mathrm{sec}^2(\theta)
$$
$$
\mathrm{cot}^2(\theta) + 1 = \mathrm{cosec}^2(\theta)
$$

### Compound Angles

Note this can be geometrically interpreted as expressing the sine of some angle $A$ increased or decreased by a factor $B$. Alternatively, it can be interpreted as expressing a sine wave of one frequency $A$ increased or decreased by $B$, as multiples of sines and cosines.

There is no point learning these by heart, can inspect when necessary.

$$
\sin(A+B) = \sin(A)\cos(B) + \cos(A)\sin(B)
$$
$$
\sin(A-B) = \sin(A)\cos(B) - \cos(A)\sin(B)
$$
$$
\cos(A+B) = \cos(A)\cos(B) - \sin(A)\sin(B)
$$
$$
\cos(A-B) = \cos(A)\cos(B) + \sin(A)\sin(B)
$$
$$
\tan(A + B) = \frac{\tan(A) + \tan(B)}{1 - \tan(A)\tan(B)}
$$
$$
\tan(A - B) = \frac{\tan(A) - \tan(B)}{1 + \tan(A)\tan(B)}
$$

### Difference of Compound Angles

This can be interpreted as adding sines and cosine functions shifted in frequency some factor $B$, expressed as multiple of sine and cosines.

$$
\sin(A+B) + \sin(A-B) = 2\sin(A)\cos(B)
$$
$$
\sin(A+B) - \sin(A-B) = 2\cos(A)\sin(B)
$$
$$
\cos(A+B) + \cos(A-B) = 2\cos(A)\cos(B)
$$
$$
\cos(A+B) - \cos(A-B) = -2\sin(A)\sin(B)
$$


### Double Angle

Geometrically, this can be interpreted as expressing a sine wave of one frequency in terms of a sine wave of half its frequency.

It is very useful to remember with these identities that we can switch between cosine and sine interpretation by halving our $\theta$, i.e.
$$
\cos(\theta) = 1 - 2\sin^2\left(\frac{\theta}{2}\right)
$$

These two are very useful:

**Cos**

\begin{gather*}
     \cos(2\theta) = 1 - 2\sin^2(\theta) \\[\16pt]
     \cos(2\theta) = 2\cos^2(\theta) - 1  \\[\16pt]
     \textnormal{from which we can derive the useful} \\[16pt]
     \cos^2(\theta) = \frac{1}{2}(1 + \cos(2\theta)) \\[16pt]
     \sin^2(\theta) = \frac{1}{2}(1 - \cos(2\theta))
\end{gather*}

This is a cool one that I always forget:

\begin{gather*}
     \cos(2\theta) = \cos^2(\theta) - \sin^2(\theta)
\end{gather*}

**Sin**
$$
\sin(2\theta) = 2\sin(\theta)\cos(\theta)
$$

**Tan**
$$
\tan(2\theta) = \frac{2\tan(\theta)}{1 - \tan^2(\theta)}
$$

## Expressing cos + sin as a shifted sine wave

This is very cool and I never really realised it. The sum of a sine and cosine of the same frequency is a shifted sine wave! We can derive the results from in the book, or think of sin and cosine as orthogonal vectors which are adding (thus modulus argument form).

Note, this is exactly how phase is determined in the Fourier transform!

\begin{gather*}
    M\sin(\theta + \alpha) = a\cos(\theta) + b\sin(\theta) \\[16pt]
    M = \sqrt{a^2 + b^2} \\[16pt]
    \alpha = \tan^{-1}\left(\frac{a}{b} \right) \\[16pt]
\end{gather*}

See modulus-argument form in complex numbers page. Note that the order of $a$ and $b$ is swapped compared to complex numbers (where sin coefficients term is on top).

# Calculus on Trig Functions

## Differentiation

Note the double angle identities are very useful when integrating and differentiating.

$$
\frac{d}{dx}\sin(x) = \cos(x)
$$
$$
\frac{d}{dx}\cos(x) = -\sin(x)
$$

For multiples of sin and cos, the double-angle identities can be very useful. e.g.

$$
\frac{d}{dx}\sin(x)\cos(x) = \dfrac{1}{2} \frac{d}{dx} \sin(2x) = \cos(2x)
$$
