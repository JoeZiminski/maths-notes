


# Argand Diagrams and Polar Coordinates

Argand diagrams are like normal graphs but the y-axis is imaginary. Important for interpreting complex numbers geometrically (e.g.modulus-argument form, roots of unity).

Complex numbers can be interpreted as vectors on the Argand diagram. Complex-magnitude form is the polar coordinate system. Complex exponential form is the phasor (vector) interpretation.

## Loci

Similar to using 2-d functions to draw shapes in the real domain (e.g. $r^2 = x^2 + y^2$ is a circle), we can use complex numbers to represent a **locus of points**. 

Interpretation-wise this is important, previously we were representing a single point, now we represent locus of points.

**Distance between points**

<p align="center">
    <img src="./arg-diagrams-polar/loci_distance.jpg" width="250">
</p>

$|z_1 - z_2| = \textnormal{distance}$. 

So like vectors, the magnitude of the difference between the vectors is the distance.

**Circle**

<p align="center">
    <img src="./arg-diagrams-polar/loci_circle.jpg" width="250">
</p>

$|z - z_1| = r$ is a circle with radius r.

Note that $z_1 = a + bi$ where a and b represent the circle center..

**Line bisector**

<p align="center">
    <img src="./arg-diagrams-polar/loci_line_bisector.jpg" width="250">
</p>

$|z - z_1| = |z - z_2|$. This makes intuitive sense, the line bisector is equal distance between $z_1$ and $z_2$.

**Half-line**

A half-line extends from a point to infinity in one direction.

<p align="center">
    <img src="./arg-diagrams-polar/loci_half_line.jpg" width="250">
</p>

$\arg(z - z_1) = \theta$ is a half line from, but not including $z_1$. This can be interpreted as all (increasing) points at a fixed angle from the real line to $z$.

## Regions

Complex number loci can be used to represent regions on the Argand diagram. This can be done by using inequality, rather than equality.

For example, for the circle $|z - z_1| \leq r$ contains all points within the circle and the border.

## Polar Coordinates

These are the same as modulus-argument form of complex numbers.

In this case, a point is represented as the magnitude and angle of a vector, $(r, \theta)$.

and the classic results hold:

\begin{gather*}
    r\cos(\theta) = x \\
    r\sin(\theta) = y \\
    r^2 = x^2 + y^2 \\
    \theta = \arctan(\frac{x}{y})
\end{gather*}

Note that sin is mapped to the imaginary axis for complex numbers because $r\sin(\theta) = y = i$

### Area enclosed by a polar curve

This is very nice, can approach as pretty much standard integration, just don't forget the square and keep notes on the bounds.

$$
Area = \int^\beta_\alpha = \frac{1}{2} r^2 d\theta
$$

Note the formula for a circle sector is $Area = \frac{1}{2}r^2 \theta$

### Tangents to polar curves

To differentiate in polar form, you can convert to parametric curve in cartesian form, then differentiate. 

Note that this indicates that parametric form is the best way to differentiate 2D curves in general.

Where $r = f(\theta)$, which is typically how polar coordinate problems are given (or, put a problem into this form):

\begin{gather*}
    x = r\cos(\theta) = f(\theta)\cos(\theta) \\
    y = r\cos(\theta) = f(\theta)\cos(\theta) \\
    \\
    \textnormal{now we can differentiate w.r.t} \ \
    \\
    \frac{dy}{dx} = \frac{dy}{d\theta} \bigg/ \frac{dx}{d\theta} 
\end{gather*}

to find the tangents, set:

\begin{gather*}
    \textnormal{parallel to the initial line: } \frac{dy}{d\theta} = 0 \\
    \\ 
    \textnormal{perpendicular to the initial line: } \frac{dx}{d\theta} = 0
    \textnormal{}
\end{gather*}
