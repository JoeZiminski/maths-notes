
# Complex Numbers and Polynomials

## Surds

A quick review of surds, important for manipulating roots. 

Summation / subtraction under the root cannot be simplified, e.g. $\sqrt{a^2 + b^2}$.

Multiplication under the root can be factored as so:

$$
\sqrt{ab} = \sqrt{a}\sqrt{b}
$$

$$
\sqrt{18} = \sqrt{9}\sqrt{2} = 3\sqrt{2}
$$

Surds can be multiplied as normal, e.g. $(4 - \sqrt{5})(3 + \sqrt{2}) = 12 - 3\sqrt{5} + 4\sqrt{2} - \sqrt{10}$

**Rationalising a denominator** is a useful trick for removing roots (or other things we want to remove, e.g. imaginary parts) from the denominator. We want to do this so we can factorise the fraction.

$$
\frac{c+d}{a+\sqrt{b}} = \frac{(c+d)(a-\sqrt{b})}{(a+\sqrt{b})(a-\sqrt{b})} = \frac{(a+\sqrt{b})(a-\sqrt{b})}{a^2-b}
$$

## Need for imaginary number - finding roots

When finding roots of polynomials, we so far restrict ourselves to the real roots i.e. when the number under the square root sign is positive. The interpretation of this is the function crosses the x-axis.

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

where the discriminant, $b^2-4ac$ determines what is under the square-root sign, and so the nature of the roots:

- $b^2-4ac > 0$: two distinct real roots
- $b^2-4ac = 0$: two equal real roots
- $b^2-4ac < 0$: two distinct imaginary roots

However, a whole class of imaginary solutions are available. In this case, the roots do not cross the x-axis. The ability of imaginary numbers to give us access to these additional roots is **amazingly important.**

As such, we define $i = \sqrt{-1}$ or $i^2 = -1$. Using surd factorisation, we can use i to find any roots. e.g. $\sqrt{-36} = \sqrt{-1}\sqrt{36} = 6i$

## Addition, Subtraction, Multiplication

Complex numbers are in the form $a + bi$. We can separate the real and imaginary parts by grouping these parts (rationalising the denominator can come in handy here). e.g. 

\begin{gather*}
    z = a+b-c+ix-iy \\ 
    z = a+b-c + i(x -y) \\
    Re\{z\} = a+b-c \\
    Im\{z\} = x-y
\end{gather*}

So we add / subtract the real and imaginary parts separately. We can multiply / divide complex numbers in the usual way (e.g. expand brackets). Interestingly, we can multiply real and imaginary parts e.g. $2(3 + 3i) = 9 + 9i$.

## Conjugation

In mathematics, a conjugate pair is defined as "a pair of binomials with identical terms by parting opposite arithmetic operators in the middle of these similar terms, e.g. $p - q$ is the conjugate of $p + q$.

For complex numbers, if $z = a + bi$, $z^* = a - bi$

Note this is useful in many circumstances, for example rationalising complex denominators.

## Finding Complex Roots up to Quartics

Form of the roots (if complex) of polynomials (don't forget the sign difference between roots and factors!). In general, if  $f(z)$ is a polynomial with real coefficients, then any imaginary roots will include a conjugate pair e.g. if $z_1$ is a root, $z_1^*$ will also be a root.


- For quadratic equations (2 roots), complex roots are always a conjugate pair \\
    e.g. if the roots are $\alpha, \beta$ we can write the polynomial as $(z - \alpha)(z-\beta) = 0$ or expanding, $z^2 - (\alpha + \beta)z + \alpha \beta = 0$

- for cubic equations (3 roots), either all three roots are real or one root is real with a complex conjugate pair as the other two roots.

- for quartic equation (4 roots), either all four roots are real, two real and two complex conjugate roots, or two separate complex conjugate pairs.

e.g. 
Show that $z^2 + 4$ is a factor of $f(z) = z^4-2z^3+21z^2-8z+68$ and hence solve the equation $z^4-2z^3+21z^2-8z+68=0$.

\begin{gather*}
    z^2=-4 \\
    z = \sqrt{-4} = \sqrt{-1}\sqrt{4} = 2i \\
    \textnormal{using factor theorem} \\
    f(2i) = 0 \\
    \textnormal{Thus this, and its conjugate, is a factor} \\
    \\
    \textnormal{Now just annoying number crunching and testing} \\
    \textnormal{e.g. seeing 4*17 = 68} \\
    (z^2+4)(z^2-2z+17) = f(z) \\
    \textnormal{Using the quadratic equation to find the last} \\
    \textnormal{2 complex roots} \\
    z = 1 \pm 4i \textnormal{ are also roots.}
\end{gather*}
