# Forms of Complex Numbers

## Standard Form

$$ z = a + bi $$

## Modulus Argument

Modulus-argument form is polar coordinates of the complex number on the Argrand plane. It consists of the vector magnitude ($|z|$) and angle ($\theta$).
$$
|z| = \sqrt{a^2 + b^2}
$$
$$
\theta = \tan^{-1} \left( \frac{b}{a} \right)
$$

so the modulus-argument form is 
$$
z = |z|(cos(\theta) + i\sin(\theta))
$$

### Multiplying and dividing complex numbers (mod-argument form)

Mod-argument form makes it easier to multiply and divide complex numbers:

\begin{gather*}
|z_1 z_2| = |z_1||z_2| \\
\textnormal{arg}(z_1 z_2) = \textnormal{arg}(z_1) + \textnormal{arg}(z_2) \\
\\
\textnormal{and for division:}
\\
\left|\frac{z_1}{z_2}\right| = \frac{|z_1|}{|z_2|} \\
\textnormal{arg}\left( \frac{z_1}{z_2} \right) = \textnormal{arg}(z_1) - \textnormal{arg}(z_2)
\end{gather*}

Prove these by expressing as $Ae^{\theta i}$.

## Euler's Relation

$$
z = Me^{i\theta} = M(\cos(\theta) + i\sin(\theta))
$$

The proof falls out of the Taylor expansion for $e^{ix}$, $cos(x)$ and $sin(x)$. See signal processing course Chapter 7 notes for proofs.

\begin{gather*}
e^{i\pi} = \cos(\pi) + i\sin(\pi) = -1 \\
\textnormal{Therefore the beautiful Euler's identity falls out of this} \\
e^{i\pi} + 1 = 0
\end{gather*}

Mutliplying and dividing:
$$
z_1 z_2 = M_1 M_2 e^{i(\theta_1 + \theta_2)}
$$

$$
\frac{z_1}{z_2} = \frac{M_1}{M_2} e^{i(\theta_1 - \theta_2)}
$$

## De Moivres Formula

### De Moivres Formula Definition and  Proof

$$
(\cos(\theta) + i\sin(\theta))^n = e^{i n \theta} = \cos(n\theta) + i\sin(n\theta)
$$

Proof by Euler's formula

$$
(e^{i\theta})^n = e^{i n \theta} = \cos(n\theta) + i\sin(n \theta)
$$
There is a nice proof by induction in cos / sine form in the Book 2 / AS.

### Expanding n-angle sine and cosines into powers expressions with De Moivre's formula

Recall the series expansion of the expression $(a + b)^n$

$$
(a + b)^n = \binom{n}{0}a^n + \binom{n}{1}a^{n-1}b + \binom{n}{2}a^{n-2}b^2 + ... + \binom{n}{n}b^n
$$

We can use the complex exponential to convert $\cos(n\theta)$ or $\sin(n\theta)$ to a power series, matching real and imaginary terms to determine their series expansion.

$$
(a + b)^n = (\cos(\theta) + i\sin(\theta))^n = \cos(n\theta) + i\sin(n\theta)
$$

## Alternate forms of $\cos(\theta)$ and $\sin(\theta)$

In complex exponential form:

\begin{gather*}
    \cos(n\theta) = \frac{e^{in\theta} + e^{-in\theta}}{2} \\
    \\
    \sin(n\theta) = \frac{e^{in\theta} - e^{-in\theta}}{2i}
\end{gather*}

Note that standard complex number form, \textbf{if} $z = \cos(\theta) + i\sin(\theta):

\begin{gather*}
2 \cos (n\theta) = z^n + \frac{1}{z^n} \\
\\ 
2i \sin(n\theta) = z^n - \frac{1}{z^n}
\end{gather*}

That come from this cute little derivation:

\begin{gather*}
    \frac{1}{z} = z^{-n} = (\cos(\theta) + i\sin(\theta)^{-n} \\
    \\ 
    = \cos(n\theta) - i\sin(n\theta) = e^{-in\theta}
\end{gather*}
so $z + z^{-1}$ the sine cancel and $z - z^{-1}$ the cosine cancel.

# Geometric series expression with complex numbers

We can use complex number is a geometric series and derive similar limiting expressions to real numbers. Indeed if $w, z \in \mathbb{C}$

\begin{gather*}
    \sum_{r=0}^{n-1} wz^r = w + wz + wz^2 + ... + wz^{n-1} = \frac{w(z^n - 1)}{z - 1}
\end{gather*}

and 

\begin{gather*}
        \sum_{r=0}^{\infty} wz^r = w + wz + wz^2 + ... = \frac{w}{1 - z}, \ |z| < 1
\end{gather*}

# Roots of Unity TODO: COMPLETELY REJIG THIS, ADD SOME EXAMPLE QUESTIONS

Now strap yourself in for one of the coolest topics in maths!

Complex numbers are different to real numbers, which can have 2 (e.g. quadratic, 4th) or 1 (cubic) roots. Complex numbers have n distinct nth roots.

$z, w \in \mathbb{C}$ and $z,w > 0$, then $z^n=w$ has $n$ distinct roots (where $n \in \mathbb{N}$).

Setting $w=1$, we can find $z^n=1$ has important geometric interpretations. It is worth thinking hard exactly what this means. For any complex number, there are $n$ distinct $nth$ roots that, when exponentiated by $n$, result in 1.

This is deeply linked to the phasor view of complex numbers, and complex numbers as rotations when expressed as complex exponentials.

Recall that due to the circularity of sine and cosine, for any complex number $z = r(\cos(\theta) + i\sin(\theta))$ you can write $z = r(\cos(\theta + 2\pi k) + i\sin(\theta + 2 \pi k))$.

**Important to remember:**

In general, the solutions to $z^n = 1$ are $z = \cos(\frac{2\pi k}{n}) + i\sin(\frac{2 \pi k}{n}) = e^{\frac{i 2 \pi k}{n}}$ FOR $k = 1, 2, ..., n$.

Note that these are all positive, but we can have negative roots too (which, due to circularity of sine and cosine, are the same as higher-positive roots (e.g. $-\frac{\pi}{3} = 2 \pi - \frac{\pi}{3} = \frac{5 \pi}{3}$)

The below features of roots of unity are very useful for solving problems. For the nth roots of unity, there is a root $\omega = e^{\frac{i2\pi}{n}}$ such that:

\begin{gather*}
    \textnormal{the nth roots are} \ 1, \omega, \omega^2, ..., \omega^{n-1} \\
    1, \omega, \omega^2, ..., \omega^{n-1}  \textbf{\textnormal{form the vertices of a regular n-gon}} \\
    1 + \omega + \omega^2 + ... + \omega^{n-1} = 0
\end{gather*}

**It is very important to think geometrically when using roots of unity}. The roots of unity are the vertices of a regular n-gon. These can also be interpreted as vectors (phasors).**
